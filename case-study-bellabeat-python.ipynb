## **4. Process Data ü§ñ**
---
Key Tasks:

* Check the data for errors.

From the above observation:

* The dataframe has 15 columns and 940 rows.
* There is no typo, NaN, and Null.
* ActivityDate is wrongly classified as object dtype and has to be converted to datetime64 dtype.
* There are 33 unique IDs, instead of 30 unique IDs that is the fitness tracker users as stated in the dataset description.

The following operation needs to be done:
* Changing the column "ActivityData" data-type to datetime64 and convert the format to yyyy-mm-dd.
* Creating new column "DayOfTheWeek" by separating the date into day of the week for further analysis.
* Creating new column "TotalMins" being the sum of "VeryActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes", and "SedentaryMinutes".
* Creating new column "TotalHours" by converting new column in column "TotalMins" to number of hours.
* Rearranging all of the columns.

```python
# Changing column ActivityDate from object to datetime64
DA_df['ActivityDate'] = pd.to_datetime(DA_df['ActivityDate'], format="%m/%d/%Y")
DA_df['ActivityDate'].head()
```

```python
# Creating new column "DayOfTheWeek" for the day of the week out of "ActivityDate"

DA_df["DayOfTheWeek"] = DA_df["ActivityDate"].dt.day_name()
DA_df["DayOfTheWeek"].head()
```

```python
# Creating new column "TotalMinutes" for the total minutes out of "VeryActiveMinutes", "FairlyActiveMinutes", "lightlyActiveMinutes", and "SedentaryMinutes"
DA_df["TotalMinutes"] = DA_df["VeryActiveMinutes"] + DA_df["FairlyActiveMinutes"] + DA_df["LightlyActiveMinutes"] + DA_df["SedentaryMinutes"]
DA_df["TotalMinutes"].head()
```

```python
# Creating new column for Total Hour 
DA_df["TotalHours"] = round(DA_df["TotalMinutes"] / 60)
DA_df["TotalHours"].head()
```

```python
# Rearranging the columns of the dataframe
new_column = ['Id', 'ActivityDate', 'DayOfTheWeek', 'TotalSteps', 'TotalDistance', 
              'TrackerDistance', 'LoggedActivitiesDistance', 'VeryActiveDistance', 
              'ModeratelyActiveDistance', 'LightActiveDistance', 'SedentaryActiveDistance', 
              'TotalHours', 'TotalMinutes', 'VeryActiveMinutes', 'FairlyActiveMinutes', 
              'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories']

DA_df = DA_df.reindex(columns = new_column)

DA_df.head(10).style.background_gradient(cmap='magma')
```

## **2. Plotting a Pie Charts to represent the percentage of users activity in minutes ‚è≤Ô∏è**
---
A pie charts is a circle-shaped graph in which a circle is divided into sectors, each representing a percentage of the whole.

```python
# Calculating the total of each columns
VeryActive = DA_df["VeryActiveMinutes"].sum()
FairlyActive = DA_df["FairlyActiveMinutes"].sum()
LightlyActive = DA_df["LightlyActiveMinutes"].sum()
Sedentary = DA_df["SedentaryMinutes"].sum()

# Plotting a pie charts
segments = [VeryActive, FairlyActive, LightlyActive, Sedentary]
labels = ["Very Active", "Fairly Active", "Lightly Active", "Sedentary"]
colors = ["cyan", "yellow", "magenta", "green"]
explode = [0.1, 0.1, 0.1, 0.2]
plt.pie(segments, labels = labels, colors = colors, startangle = 180,
        wedgeprops = {"edgecolor": "black"}, shadow = True,
        explode = explode, autopct = "%1.1f%%")
plt.title("Percentage of Users' Activities in Minutes")
plt.legend(title = "Activities", loc = "lower right")
plt.tight_layout()
plt.show()
```

Key findings from the pie charts above:
* The pie charts indicates that users were using the FitBit app to log for their daily activities. 
* Sedentary state of users make up the most (81.33%) of other activities in minutes.
* The pie charts could indicate that the users did not use the app as intended such as working out, running, fitness, etc. Despite the fact that the app intention is to encourage fitness of the users.

## **3. Plotting a scatter plot to show correlations between two variables üìä**
---
A scatterplot depicts the magnitude, direction, and shape of a relationship between two quantitative variables.

```python
# Initialise the subplot function using number of rows and columns
fig, axis = plt.subplots(1, 2, figsize=(15, 8))

m1, b1 = np.polyfit(DA_df.TotalSteps, DA_df.Calories, 1)
mean_calories = DA_df.Calories.mean()
mean_steps = DA_df.TotalSteps.mean()

m2, b2 = np.polyfit(DA_df.TotalDistance, DA_df.Calories, 1)
mean_calories = DA_df.Calories.mean()
mean_distance = DA_df.TotalDistance.mean()

# Plotting a scatter plot
axis[0].scatter(DA_df.TotalSteps, DA_df.Calories, alpha = 0.8, c = DA_df.Calories, cmap = "hot")
axis[0].plot(DA_df.TotalSteps, m1 * DA_df.TotalSteps + b1, '-', color='brown')

# Adding annotations and visuals
# axis[0].colorbar(orientation = "vertical")
axis[0].axvline(mean_steps, color = "Green", label = "Mean Total Steps")
axis[0].axhline(mean_calories, color = "Blue", label = "Mean Calories")
axis[0].set_xlabel("Total Steps")
axis[0].set_ylabel("Calories burned")
axis[0].set_title("Correlation Between the Total Steps and Calories Burned")
axis[0].grid(True)
axis[0].legend(loc = "lower right")

# Plotting a scatter plot
axis[1].scatter(DA_df.TotalDistance, DA_df.Calories, alpha = 0.8, c = DA_df.Calories, cmap = "hot")
axis[1].plot(DA_df.TotalDistance, m2 * DA_df.TotalDistance + b2, '-', color='brown')

# Adding annotations and visuals
axis[1].axvline(mean_distance, color = "Green", label = "Mean Total Distance")
axis[1].axhline(mean_calories, color = "Blue", label = "Mean Calories")
axis[1].set_xlabel("Total Distance")
axis[1].set_ylabel("Calories burned")
axis[1].set_title("Correlation Between the Total Distance and Calories Burned")
axis[1].grid(True)
axis[1].legend(loc = "lower right")
#plt.colorbar(orientation = "vertical")
plt.show()
```

Key findings from the scatter plots above:
* There is a positive correlation as shown in a brown line in both graphs between the total steps and the calories burned and between total distance and calories burned.
* There are outliers in this data which can be caused by errors in data collection, miscalculations, data contamination, human error, etc.

```python
m3, b3 = np.polyfit(DA_df.TotalHours, DA_df.Calories, 1)
mean_calories = DA_df.Calories.mean()
mean_sedentary = DA_df.SedentaryMinutes.mean()/60.0
mean_hours = DA_df.TotalHours.mean()

# Plotting a scatter plot
plt.figure(figsize=(8,6)) # Specify size of the chart
plt.scatter(DA_df.TotalHours, DA_df.Calories, alpha = 0.8, c = DA_df.Calories, cmap = "hot")
plt.plot(DA_df.TotalHours, m3 * DA_df.TotalHours + b3, '-', color='brown')

# Adding annotations and visuals
plt.colorbar(orientation = "vertical")
plt.axhline(mean_calories, color = "Blue", label = "Mean Calories")
plt.axvline(mean_sedentary, color = "Green", label = "Mean Sedentary")
plt.axvline(mean_hours, color = "Purple", label = "Mean Hours")
plt.xlabel("Hours Logged")
plt.ylabel("Calories Burned")
plt.title("The Correlation Between Calories Burned for Every Hour Logged")
plt.grid(True)
plt.legend()
plt.show()
```

Key findings from the scatter plot above:
* A weak positive correlation as shown in a brown line whereas the increase of the hours logged does not translate to more calories being burned. That is largely due to the average sedentary hours (green line) plotted at the 16 to 17 hours range.
* Some outliers in the data
